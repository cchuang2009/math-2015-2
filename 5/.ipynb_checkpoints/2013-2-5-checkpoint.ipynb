##Animation

   **"Frame by frame"** is the basic technique of making animation. The technique involves taking a sequence of shoots on something and combine them into a **movie**.
For example, "director" shoots one frame of graphic output.
Then the graphic output varies a tiny bit and another frame is shot. This process is repeated over and over until the full output
is completed. Usually, one second of stop-motion animation requires 24 individual frames, so called 24 *fps*. Afterwards, the frames are put together
and played one after another, creating the dynamical appearance of the computation we made.
<br>

   Such work could be easily implemented by scitools module, especially **movie(...)** function. Here is the example for making *phase diagram* of *Van der Pol Oscillator*. There are two kinds of animation available here, *gif* and *webm*.
The former animation is created by *ImagMagick/convet* and the latter is created by *ffmpeg* and could be directed played by HTML5's **&lt;video controls ... &gt;** which is supported almost by recent browsers, *google's chrome, firefox* for instance.
<p>
Example:

**$a^\circ$ ) 3D picture for $f(x,y)=x^4+y^4-c x y$**
    <br>
    At first as $c=0$, $f(x,y)$ would attain its minimum at the critical point $(x_0,y_0)=(0,0)$; but $f(x,y)$ would change to be a saddle point while $c$ increases from $0$.

**$b^\circ) $ Motion of Van der Burger Equation**

\begin{aligned} 
\frac{d^2 x}{d t^2}-\mu(1-x^2)\frac{d x}{d t} +x&=&0 \cr
y&=&\frac{d x}{d t} \cr
\text{Iniatial condition:}&& x_0^2+y_0^2=1
\end{aligned} 

```python
import numpy as np
from scitools.easyviz import *
import glob, os

""" delete all the pictures if created ever """
for filename in glob.glob('tmp*.eps'):
    os.remove(filename)

u=1.0
t0=0
s=1.0
L=2.*np.pi
n=36.
dx=L/n
dt=0.01

x0=np.asarray([2.*np.cos(k*dx) for k in np.arange(0,n+1)])
y0=np.asarray([2.*np.sin(k*dx) for k in np.arange(0,n+1)])

xsol=x0
ysol=y0
f_no=21

for i in arange(0,f_no):
 """ Main computation algorithm"""   
 xsol=x0+dt*y0
 ysol=y0+dt*(-x0+y0*(1-x0*x0))
 x0=xsol
 y0=ysol
    
 plot(xsol,ysol,'r-6',axis=(-4, 4, -4, 4), aspect=True, size=[4,4], xlabel='X', ylabel='Y', legend='s=%4d' % i, \
          hardcopy='tmp%04d.eps' % i);

""" Generate GIF Figure"""
movie('tmp*.eps',encoder='convert', fps=4, quiet=True, output_file='vanderpol.gif')
```

# Arnold's Cat

Look at the Ad of opera, **Cats**, the first frame of movie below.




Russian Mathematician, **V.I. Arnold** ever considered a picture, Cat,  translated by the following relationship:


\begin{align}
x&=&2x+y \quad(\mod 1), \cr
y&=&x+y \quad(\mod 1) 
\end{align}
where  $(x,y)$ is the coordinates in a picture. The translation rule is equivalent to
the linear transformation:
$$
{\left(\begin{matrix}
x_{n+1} \cr
y_{n+1}
\end{matrix}\right)
}={
\left(\begin{matrix}
2 &1 \cr
1 & 1
\end{matrix}
\right)
}
{
{\left(
\begin{matrix}
x_n \cr
y_n 
\end{matrix}
\right)
} {(\mod 1)}
}
$$
under the assumption of the size of picture being set as $1\times1$. The following example uses **Python Image Library, PIL**, to manipulate linear transformation of the image and create the movie clip in *webm* format by **scitools**.

```python
import Image 
import ImageChops
import ImageFont
import ImageDraw

n=100
# checks if two images are equal
def equal(im1, im2): 
    return ImageChops.difference(im1, im2).getbbox() is None
 
# add a text caption to an image (adds space at the bottom)
def addCaption(im, text, pointSize):
    size = im.size
    draw = ImageDraw.Draw(im)
 
    textFontLocation = "/usr/share/fonts/TTF/DejaVuSans.ttf"
    topBottomMargin = 5
 
    textFont = ImageFont.truetype(textFontLocation ,pointSize)
    textSize = draw.textsize(text, font=textFont) # the size of the text box!
 
    newImage = Image.new(im.mode,\
            (im.size[0], im.size[1]+textSize[1]+2*topBottomMargin), "white" )
    newImage.paste( im, (0,0) )
    draw = ImageDraw.Draw(newImage)
 
    textX = (newImage.size[0] / 2.0) - (textSize[0] / 2.0)
    textY = newImage.size[1] - textSize[1] - topBottomMargin
 
    draw.text((textX, textY), text, fill="black", font=textFont)
    return newImage
 
# add a solid border to an image
def addBorder(im, color, thickness):
    newImage = Image.new(im.mode,\
            (im.size[0]+thickness*2, im.size[1]+thickness*2), color)
    newImage.paste( im, (1,1) )
    return newImage
 
# scale an image up or down (doesn't resample, so the pixels are clear)
def scaleImage(im, factor):
    newImage = Image.new(im.mode,\
            (im.size[0]*factor, im.size[1]*factor), "white" )
    newImage.paste( im.transform((im.size[0]*factor, im.size[1]*factor),\
            Image.AFFINE,\
            (1/float(factor),0,0,0,1/float(factor),0) ),\
            (0,0) )
    return newImage
 
def saveImages(im, iteration):
    im.save("Catmap%04d.png"%iteration) #save the simple image
 
    #resize image, add border and caption
    imageLarge = scaleImage(im, 3)
    imageLarge = addBorder(imageLarge, "black", 1)
    imageLarge = addCaption(imageLarge, str(iteration), 25)
    imageLarge.save("CatmapLargeNumbered%04d.png"%iteration)
 
# CONTROL STARTS HERE
 
inFile = "Cat.png" #input image
 
image0 = Image.open(inFile, 'r')

imageOrig = image0.copy() # keep to original image to see when we return to it
 
image1 = Image.new(image0.mode, (n,n)) # temp image to copy pixels to
pixels1 = image1.load()
 
# BEGIN THE CAT MAPPING PROCESS
 
iteration = 0
saveImages(image0, iteration) # save the first image
while True:
    pixels0 = image0.load() #reload the iterated image
 
    for x in range(n):# perform the mapping
        for y in range(n):
            newX = (2*x + y) % n #find new location
            newY = (x + y) % n
            pixels1[newX, newY] = pixels0[x, y] # copy the pixel over
 
    image0 = image1.copy() #transfer back to image0 for the next iteration
    iteration += 1
    saveImages(image0, iteration) # save this iteration's image
 
    if equal(image0, imageOrig):
        break
```

###Example 1. (1D Heat Equation)

The **Intial Value Problem (IVP)**:
\begin{align}
\frac{\partial u}{\partial t} &=& \frac{\partial^2 u}{\partial x^2} \cr
u(0,x) &=& f(x)
\end{align} 
The analytic solution is $u(t,x)=e^{-t^2}f(x)$.

```python
import numpy as np
from scitools.easyviz import *
from scitools.all import *
import glob, os

""" delete all the pictures if created ever """
for filename in glob.glob('tmp*.eps'):
    os.remove(filename)

nt=20
nx=50
t=0
x0=0.
xf=np.pi*2

x=np.linspace(x0,xf,nx+1)

dt=1/float(nt)
for i in arange(0,nt):
    u=np.exp(-i*i*dt*dt)*np.sin(x)
    plot(x,u,'r-6',axis=(0, 2*np.pi, -1, 1), aspect=True, size=[12.3,4], legend='Time=%f' % float(i*dt), \
          hardcopy='tmp%04d.eps' % i);

""" Generate GIF Figure"""
movie('tmp*.eps',encoder='convert', fps=4, quiet=True, output_file='heat1D.gif')
```

## Numeric Scheme for Heat Equation:
###Schemes for Differentiation:
*change with respect to time, $t$;$$\frac{\partial u}{\partial t} \sim\frac{u(x_i,t^{j+1})-u(x_i,t^j)}{\vartriangle t}=\frac{u_i^{j+1}-u_i^j}{\vartriangle t} $$
*change with respect to space, $x$:
$$\frac{\partial u}{\partial x} \sim \frac{u(x_{i+1/2},t^j)-u(x_{i-1/2},t^j)}{\vartriangle x}$$
$$\frac{\partial^2 u}{\partial x^2} \sim \frac{\frac{\partial u}{\partial x}(x_{i+1/2},t^j)-\frac{\partial u}{\partial x}(x_{i-1/2},t^j)}{\vartriangle x}$$
$$\frac{\partial^2 u}{\partial x^2} \sim\frac{u(x_{i+1},t^j)-2u(x_i,t^j)+u(x_{i-1},t^j)}{ \vartriangle x^2}=\frac{u^j_{i+1}-2u_i^j+u^j_{i-1}}{\vartriangle x^2} $$
<p>
This implies that $u_.^{j+1}$ can be evaluated by the previous state $u_.^j$:
    $$u_i^{j+1}=u_i^j+\frac{\vartriangle t}{\vartriangle x^2}(u^j_{i+1}-2u_i^j+u^j_{i-1})$$
<p>
    The initial condition, $u(x,0)=\sin(x),0\le x\le 2\pi$, assigns the following rule:
    $$u_i^0=u(x_i,0)=\sin(x_i)=u(i\vartriangle x), i=0,1,\cdots,n  $$
    where $[0,2\pi]$ is equally-like divided into $n-$subintervals as follows:
    $$0=x_0\le x_1\le x_2\le \cdots \le x_{n-1}\le x_n=2\pi \text{ and }x_i-x_{i-1}=\vartriangle x=\frac{2\pi}{n}$$

```python
import numpy as np
from scitools.easyviz import *
from scitools.all import *

nt=40
nx=20
t0=0
x0=0.
xf=np.pi*2
tf=1

dt=(tf-t0)/float(nt)
dx=(xf-x0)/float(nx)

x=np.linspace(x0,xf,nx+1)
u=zeros(nx+1)
SOL=[]

for i in arange(0,nx+1):
    u[i]=np.sin(x[i])
SOL.append(u)
plot(SOL[0])
```

```python
t=0
dxdt=dt/dx/dx
up=zeros(nx+1)

while t<=tf:
      t  += dt;
      up[1:nx] = u[1:nx]+dxdt*(u[2:nx+1]-2*u[1:nx]+u[0:nx-1])
      u=up
      SOL.append(up)
```

#Reaction-diffusion Equation

A typical *Reaction-diffusion equation* is as the following form:
\begin{eqnarray*}
  \frac{\partial V}{\partial t} & = & \quad\nabla^2 V \quad+ f (V)\cr
  &  & \qquad\uparrow \qquad\quad\uparrow\cr
  &  & \text{diffusion} \quad \text{reaction}
\end{eqnarray*}

##Chaotic Electric Waves

R.R Aliev and A.V. Panfilov, *A simple two-variable model of cardiac excitation*, **Chaos, Solitons and Fractals**, 1996. 

\begin{aligned}
e_t &=& \delta\vartriangle e -k e(e-a)(e-1)-er \cr
r_t &=& -\left[\epsilon+\frac{\mu_1 r}{\mu_2+e}\right][r+k e(e-b-1)]
\end{aligned}
Boundary Conditions: $e(x,y,t)=r(x,y,t)=0$ for $(x,y)\in\partial \Omega$.

Initial Conditions: $e(x,y,0)=e_0(x,y)$ and $r(x,y,0)=r_0(x,y)$ 

<p>
    Parameters used here:
    $\mu_1=0.2,\mu_2=1.3,k=8,\epsilon=0.01,a=0.12,b=0.1, \delta=0.005$<br>
    Grids of space: $x_i=i*h,i=0,1,2,\cdots$, $y_j=j*k,j=0,1,2,\cdots$, $h=k=1/(M+1)$ and $M=128$
<br>=====================<br>

###Numeric Scheme
* $ e_{i,j}^n$: $e(x_i,y_j)$ at time $t_n$;
* $ r_{i,j}^n$: $r(x_i,y_j)$ at time $t_n$;    
    
\begin{aligned}
    \frac{ e_{i,j}^{n+1}-e^n_{i,j} }{ \vartriangle t } &=&  \delta\frac{e^n_{i+1,j}-2e^n_{i,j}+e^n_{i-1,j}}{h^2} +\delta\frac{e^n_{i,j+1}-2e^n_{i,j}+e^n_{i,j-1}}{h^2} \cr
                                                      && -k e^n_{i,j} (e^n_{i,j}-a) (e^n_{i,j}-1)-e^n_{i,j}r^n_{i,j} \cr
 \frac{ r_{i,j}^{n+1}-r^n_{i,j} }{ \vartriangle t} &=& -\left( \epsilon+ \frac{\mu_1  r^n_{i,j}}{\mu_2 +e^n_{i,j}}\right)\left(r^n_{i,j}+k e^n_{i,j}(e^n_{i,j}-b-1)\right) \cr
 \Rightarrow   e_{i,j}^{n+1} &=& \rho (e^n_{i+1,j}+e^n_{i-1,j}+e^n_{i,j+1}+e^n_{i,j-1})-(1-4\rho)e^n_{i,j}\cr
                    && -\vartriangle t \left[ k e^n_{i,j} (e^n_{i,j}-a) (e^n_{i,j}-1)-e^n_{i,j}r^n_{i,j} \right] \cr
               r_{i,j}^{n+1} &=& r^n_{i,j} -\vartriangle t \left( \epsilon+ \frac{\mu_1  r^n_{i,j}}{\mu_2 +e^n_{i,j}}\right)\left(r^n_{i,j}+k e^n_{i,j}(e^n_{i,j}-b-1)\right) 
\end{aligned}
    where $\rho=\delta \frac{\vartriangle t} {h^2}$

```python
from scitools.all import *
from scitools.StringFunction import StringFunction
import sys
my1 = 0.2
my2 = 1.3
k = 8.
epsilon = 0.01
a = 0.12
b = 0.1
delta = 0.0005
N = 128
M = 128
e0 = '(1-x)**4.*(1-y)**4.'
r0 = '1'
e0 = StringFunction(e0, independent_variables=('x', 'y'))
r0 = StringFunction(r0, independent_variables=('x', 'y'))
e = zeros([M + 2, M + 2])
em = e.copy()
r = zeros([M + 2, M + 2])
rm = r.copy()
x = linspace(0, 1, M + 2)
y = linspace(0, 1, M + 2)
T = 500
t = 0
h = 1./(M + 1)
dt = 1/float(N)
rho = delta*dt/h**2

for i in range(len(x)):
    for j in range(len(y)):
        em[i][j] = e0(x[i], y[j])
        rm[i][j] = r0(x[i], y[j])
t += dt
while t <= T:
   if ((t%10.) < dt):
      # Plot every 10 steps
      contourf(em,size=[4,4], axis='tight')
      hardcopy("base%08.3f.png" %t, dpi=75)
   # First, add inner points
   em[1:M+1, 1:M+1] += rho*(em[0:M, 1:M+1] + em[2:M+2, 1:M+1] +
                            em[1:M+1, 0:M] + em[1:M+1, 2:M+2] -
                            4*em[1:M+1, 1:M+1])
   # Secondly, add all points
   e = em - dt*(k*em*(em - a)*(em - 1) + em*rm)
   r = rm - dt*(epsilon + my1*r/(e + my2))*(r + k*e*(e - b - 1))
   em = e.copy()
   rm = r.copy()
   t += dt
```

##FitzHugh-Nagumo Models:
\begin{eqnarray*}
  \frac{\partial V}{\partial t} & = & \nabla^2 V + \frac{1}{\epsilon} \left( V
  - V^3 / 3 - W \right)\cr
  \frac{\partial W}{\partial t} & = & \epsilon \left( V - \gamma W + \beta
  \right)\cr
  \epsilon, \gamma, \beta & = & 0.2, 0.6, 0.7
\end{eqnarray*}
where $V$ is transmembrance potential and $W$ is recovery variable. When $\beta$ is larger than $\beta_c=7/14$, the model will become
unstable.<br />
**Ref:**<a href="http://pages.physics.cornell.edu/~myers/teaching/ComputationalMethods">Computational Methods for Nonlinear Systems</a>

```python
import math
import Tkinter 
import Image, ImageTk

DefaultImageSize = (400,400)

root = Tkinter.Tk()
root.withdraw()

class DynamicLattice (Tkinter.Label):

    def __init__(self, shape, size=DefaultImageSize, mode='P', 
                 zmin=0.0, zmax=1.0):
        top = Tkinter.Toplevel()
        top.title('DynamicLattice')
        Tkinter.Label.__init__(self, top)
	self.shape = shape
	self.size = size
	self.mode = mode
	self.zmin = zmin
	self.zmax = zmax
	self.zrange = zmax-zmin
        self.canvas = Tkinter.Canvas(top, 
				     width=self.size[0], height=self.size[1])
        Tkinter.Widget.bind(self.canvas, "<Button-1>",
                           self.mouseDown)
        Tkinter.Widget.bind(self.canvas, "<Button1-ButtonRelease>",
                           self.mouseUp)
        self.mousecoords = []
	self.im = Image.new(self.mode, self.shape)
	self.displayIm = self.im.resize(self.size)
	if self.mode == '1':
	    self.tkimage = \
			 ImageTk.BitmapImage(self.displayIm,
					     foreground="white")
	else:
	    self.tkimage = \
			 ImageTk.PhotoImage(self.displayIm)
	self.canvas.create_image(0, 0, anchor=Tkinter.NW, image=self.tkimage)
        self.canvas.pack()

    def setTitle(self, title):
        self.master.title(title)

    def mouseDown(self, event):
        x0 = self.canvas.canvasx(event.x)
        y0 = self.canvas.canvasx(event.y)
        sx0 = int(self.shape[0] * float(x0)/self.size[0])
        sy0 = int(self.shape[1] * float(y0)/self.size[1])
        self.mousecoords = [sx0, sy0]

    def mouseUp(self, event):
        sx0, sy0 = self.mousecoords
        x1 = self.canvas.canvasx(event.x)
        y1 = self.canvas.canvasx(event.y)
        sx1 = int(self.shape[0] * float(x1)/self.size[0])
        sy1 = int(self.shape[1] * float(y1)/self.size[1])
        X0, X1 = min(sx0, sx1), max(sx0, sx1)
        Y0, Y1 = min(sy0, sy1), max(sy0, sy1)
        self.mousecoords = [X0, Y0, X1, Y1]
    
    def IsBoxSelected(self):
        return len(self.mousecoords)==4

    def GetMouseBox(self):
        mc = self.mousecoords[:]
        self.mousecoords = []
        return mc

    def display(self, array, site=None):
        if site is not None:
            self.im.putpixel(site, self.grayscale(array[site]))
	else:
	    for i in range(array.shape[0]):
                for j in range(array.shape[1]):
                    self.im.putpixel((i,j), self.grayscale(array[i,j]))
	self.displayIm = self.im.resize(self.size)
        self.tkimage.paste(self.displayIm)
        self.canvas.update()

    def grayscale(self, value):
        sval = (value-self.zmin)/(self.zrange)
        if sval < 0.: sval = 0.
        if sval > 1.: sval = 1.
        return int(255.*(sval))

    def set_zmin(self, zmin):
        self.zmin = zmin
        self.zrange = self.zmax - self.zmin
        
    def set_zmax(self, zmax):
        self.zmax = zmax
        self.zrange = self.zmax - self.zmin
```

```python
import scipy, scipy.optimize

#from DynamicLattice import *


DISPLAYINTERVAL=10

def del2_5(A, dx):
    nx, ny = A.shape
    del2 = scipy.zeros((nx, ny), float)
    del2[1:-1, 1:-1] = (A[1:-1,2:] + A[1:-1,:-2] + \
                A[2:,1:-1] + A[:-2,1:-1] -4.*A[1:-1,1:-1])/(dx*dx)
    return del2 

def del2_9(A, dx):
    nx, ny = A.shape
    del2 = scipy.zeros((nx, ny), float)
    del2[1:-1, 1:-1] = ((2./3.)*A[1:-1,2:] + (2./3.)*A[1:-1,:-2] + \
                       (2./3.)*A[2:,1:-1] + (2./3.)*A[:-2,1:-1] - \
                       (10./3.)*A[1:-1,1:-1] + \
                       (1./6.)*A[2:,2:] + (1./6.)*A[2:,:-2] + \
                       (1./6.)*A[:-2,2:] + (1./6.)*A[:-2,:-2])/(dx*dx)
    return del2 

del2 = del2_9

def FindFixedPoint(gamma, beta):
    f = lambda v, gamma, beta: (v-(v**3)/3.)-((1./gamma)*(v+beta))
    Vstar = scipy.optimize.brentq(f, -2., 2., args=(gamma, beta))
    Wstar = ((1./gamma)*(Vstar+beta))
    return Vstar, Wstar

def CopyGhost(A):
    A[0,:] = A[1,:]
    A[:,0] = A[:,1]
    A[-1,:] = A[-2,:]
    A[:,-1] = A[:,-2]

class FitzNag2D:
      def __init__(self, N=100, dx=1.0,eps=0.2, gamma=0.8, beta=0.4):
          self.DL = DynamicLattice((N, N), zmin=-2.0, zmax=2.0)
          self.N = N
          self.dx = dx
          self.L = N*dx
          self.eps = eps
          self.gamma = gamma
          self.beta = beta
          self.pulseheight = 3.0
          Vstar, Wstar = FindFixedPoint(gamma, beta)
          self.V = Vstar*scipy.ones((N, N), float)
          self.W = Wstar*scipy.ones((N, N), float)
          self.V[(N/2-5):(N/2+5), (N/2-5):(N/2+5)] = self.pulseheight
          self.t = 0.
      def rhs(self):
          self.dV_dt = del2(self.V, self.dx) + \
                       (1./self.eps) * (self.V - (1./3.)*self.V**3 - self.W)
          self.dW_dt = self.eps*(self.V - self.gamma*self.W +self.beta)

      def step(self, dt):
          CopyGhost(self.V)
          self.rhs()
          self.V += dt * self.dV_dt
          self.W += dt * self.dW_dt
      def run(self, T, dt):
         count=0
         t0 = self.t
         while self.t < t0 + T:
               if count%DISPLAYINTERVAL==0:
                  if self.DL.IsBoxSelected():
                     x0, y0, x1, y1 = self.DL.GetMouseBox()
                     self.V[x0:x1, y0:y1] = self.pulseheight
                  self.DL.setTitle('t = %f' % self.t)
                  self.DL.display(self.V)
               self.step(dt)
               self.t += dt
               count +1

def demo():
    print "Cardiac Dynamics:"
    print "Fitzhugh Nagumo in 2D"
    print "Click and drag a rectangle for electroshock"

    f = FitzNag2D(N=100, dx=1.)
    f.run(200., 0.1)
    return f

if __name__ == '__main__':
   f = demo()                                   
```

<p>Generally, a baseball in flight is subject to three distinct forces:</p>
<ul>
<li><em>gravity</em>, which causes the ball to accelerate vertically downwards at $g=9.8 m/s^{-2}$;</li>
<li><em>air drag</em>, which impedes the ball's motion through the air;</li>
<li><em>Magnus force</em>, which permits the ball to curve laterally.</li>
</ul>
<p>Stoke's Law:</p>
<p>an object which moves&nbsp; through a viscous fluid is directly <em>proportional</em> to the speed   of that object with respect to the fluid.</p>
<p>And a sphere of radius  $r$, moving with speed $v$ through a fluid  whose coefficient of viscosity is $\eta$,  experiences a drag force  given by Stokes' law: $$ F_D=6\pi \eta r v$$</p>
<p>At velocities sufficiently low for Stokes' law to be valid,  air is  able to flow smoothly around a passing projectile. However, at higher  velocities, the projectile's motion is too rapid for this to occur.  Instead, the projectile  effectively knocks the air out of its way. The  total mass of air which the projectile comes into contact with per  second is $\rho\,v\,A$, where $\rho$ is the air density, $v$ the  projectile speed, and $A$ the projectile's cross-sectional area.    Suppose, as seems reasonable, that the projec\frac{d x}{d t}tile imparts  to this air mass a speed $v'$ which is directly proportional to $v$.  The rate of momentum gain of the air, which is equal to the drag force  acting on the projectile, is approximately</p>
<p>$$ F_D=\frac{1}{2}C_D(v)\rho A v^2$$</p>
<p>where the <em>drag coefficient</em>, $C_D(v)$, is a dimensionless quantity.</p>
<p>&nbsp;</p>
<p>Giordano has developed the following useful formula which quantifies the drag force acting on a baseball:</p>
<p>$$ \frac{{\bf f}_D}{m} = -F(v)\,v\,{\bf v}, $$</p>
<p>where</p>
<p>$$ F(v) = 0.0039 + \frac{0.0058}{1+{\rm exp}[(v-v_d)/{\mit\Delta}]}. $$</p>
<p>Here,  $v_d = 35\,{\rm m/s}$ and  ${\mit\Delta} = 5\,{\rm m/s}$.</p>
<p>&nbsp;</p>
<p>If we think of drag forces as exerting a sort of pressure on the ball  then we can readily appreciate that when the unequal drag forces acting  on the ball's upper and lower surfaces are added together there is a  component of the resultant force acting <em>upwards</em>. This force--which is known\frac{d x}{d t} as the <em>Magnus force</em>,  after the German physicist Heinrich&nbsp;Magnus, who first described it in  1853--is the dominant spin-dependent force acting on baseballs. The  Magnus force can be written</p>
<p>$$ F_M = S(v) \omega \times v, $$</p>
<p>where  $\omega$ is the  angular velocity vector  of the ball.  According to Adair and Giordano, it is a fairly good approximation to  take</p>
<p>$$ B= \frac{S}{m} = 4.1\times 10^{-4} $$</p>
<p>for baseballs. Note that $B$  is a dimensionless quantity. The  magnitude of the Magnus force is about one third of the force due to  gravity for typical curveballs.</p>
<p>The equations of motion of a baseball can be written as the following set of coupled first-order o.d.e.s:</p>
<p>$$   (\frac{d x}{d t},\frac{d y}{d t},\frac{d z}{d t}) = (v_x,v_y,v_z) $$</p>
<p>&nbsp;</p>
<p>$$ \frac{d v_y}{d t}=- F(v) v v_y + B \omega v_x \cos\phi,$$</p>
<p>&nbsp;</p>
<p>$$\frac{d v_z}{d t}=-g- F(v) v v_z - B \omega v_x \sin\phi$$</p>
<p>with initial conditions:</p>
<p>$$(x_0,y_0,z_0,v_{x}(0),v_{x}(0),v_{x}(0))=(0,0,0,v_0\cos\theta,0,v_0\sin\theta)$$</p>
<p>&nbsp;</p>
<p>where $v_0$ is the initial speed of the pitch, and $\theta$ is its initial angle of elevation.</p>
