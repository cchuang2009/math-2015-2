# Just for fun, ran the Speech through a Markov chain
Markov chain code inspired from [here](http://pythonadventures.wordpress.com/2014/01/23/generating-pseudo-random-text-using-markov-chains/)

How we will do
---    
    
1. speech is splited into sequence of element with consective three words;

2. use the first two words as key (index) and the left as value, i.e

```
[first, second, third] --> element['first,second'] = third
   
```    

3. If not end of sentence, one of '.' , '?', '!',  create the article by assembling the elemnts as follow:

```
        element['first,second'], element['second,third'], ...
value:           third                    fourth

-> 
   first second third fourth ...
```   
   

````python
from random import choice
 
EOS = ['.', '?', '!']
 
def build_dict(words):
    """
    Build a dictionary from the words.
 
    (word1, word2) => [w1, w2, ...]  # key: tuple; value: list
    """
    d = {}
    for i, word in enumerate(words):
        try:
            first, second, third = words[i], words[i+1], words[i+2]
        except IndexError:
            break
        key = (first, second)
        if key not in d:
            d[key] = []
        d[key].append(third)
 
    return d
 
 
def generate_sentence(d):
    temp = [key for key in d.keys() if key[0][0].isupper()]
    key = choice(temp)
 
    wli = []
    first, second = key
    wli.append(first)
    wli.append(second)
    while True:
        try:
            third = choice(d[key])
        except KeyError:
            break
        wli.append(third)
        if third[-1] in EOS:
            break
        key = (second, third)
        first, second = key
 
    return ' '.join(wli)
````

####Build the markov chain with the words in the speech

````python
words=speech.split()
d = build_dict(words)
````

#### Generate a random one liner text using the Markov Chain

````python
sent = generate_sentence(d)
print(sent)
````

#### Generate a random paragraph using the Markov Chain

````python
text=""
for i in range(10):
     sent = generate_sentence(d)
     text+=sent+" "
print(text)
````
