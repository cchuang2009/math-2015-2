Statistic for testing association between $A$ and $B$:
$$ \chi^2_{(m-1)(n-1)}=\sum_{i,j}^{m,n}\frac{(O_{i,j}-E_{i,j})^2}{E_{i,j}}$$
2.  The odds ratio (OR): investigation of the nature or direction of the association,
$$\hbox{OR}= \frac
      {{\cal P}(\hbox{Disease | exposed})/[1-{\cal P}(\hbox{Disease | exposed}]}
      {{\cal P}(\hbox{Disease | unexposed})/[1-{\cal P}(\hbox{Disease | unexposed}]}      
$$
 </font></big>
  a) Statistics for $2\times2$ contingency table:
---
$$\widehat{\hbox{OR}}=\frac{O_{11}O_{22}}{O_{12}O_{21}}$$
is not **Normal distributed**; but its logarithm is:
$$Z=\frac{\ln\left(\frac{O_{11}O_{22}}{O_{12}O_{21}}\right)}
    {\sqrt{1/O_{11}+1/O_{12}+1/O_{21}+1/O_{22}}}$$
  b)   Confidence intervals: a $(1 âˆ’ \alpha)$% two-sided confidence interval of the log-odds ratio is given by
$$\left(\ln(\overline{OR})-z_{1-\alpha/2}\sqrt{1/O_{11}+1/O_{12}+1/O_{21}+1/O_{22}},
  \ln(\overline{OR}+z_{1-\alpha/2}\sqrt{1/O_{11}+1/O_{12}+1/O_{21}+1/O_{22}}
  \right)$$
  and the configence interval is 
$$\left(\exp\left\{\ln(\overline{OR})-z_{1-\alpha/2}\sqrt{1/O_{11}+1/O_{12}+1/O_{21}+1/O_{22}}\right\},
  \exp\left\{\ln(\overline{OR}+z_{1-\alpha/2}\sqrt{1/O_{11}+1/O_{12}+1/O_{21}+1/O_{22}}\right\}
  \right)$$ 
3.  Combining $2\times2$ contingency tables:
  
  a)  test statistic:
$$ X^2=\sum_{i=1}^2 w_i(y_i-Y)^2\sim\chi^2_1$$
where $y_i=\ln(\widehat{OR_i})=\ln(a_id_i/(b_ic_i))$, $w_i=1/(1/a_i+1/b_i+1/c_i+1/d_i)$ and 
$Y=\sum_iw_iy_i/\sum_iw_i$.

If we want to test whether the proportion of unprotected cyclists that have serious head injuries is higher than that of protected cyclists, we can carry out a general test of hypothesis involving the two proportions $p_1 = 17/147 = .115$, and $p_2 = 218/646 = .337$.

Test Hypothesis
---
By two-sample test, design the test as follows:

$H_0 : p_1=p_2$<br>
$Ha : p_1\ne p_2$

- Statistics: $\hat p_1\sim$ Binomial($n_1=147,p_1=.115$),  $\hat p_2\sim$ Binomial($n_2=646,p_2=.337$)
$$ \frac{\hat p_1-\hat p_2-(p_1-p_2)}{\sqrt{\hat p(1-\hat p)(1/n_1+1/n_2)}}\sim N(0,1)$$
- if $H_0$ with $\alpha=0.05$: $\hat p=235/793$
\begin{eqnarray}
 \frac{\hat p_1-\hat p_2}{\sqrt{\hat p(1-\hat p)(1/n_1+1/n_2)}}&=&\frac{.115-.337}{\sqrt{.296(1-.296)
      (1/147+1/646)}}=-5.32\le-1.96=z_{0.0255}\\
\end{eqnarray}
reject $H_0$.

$\chi^2$-test 
---
Now suppose that you wanted to determine whether there is any *association* between wearing helmets and frequency of brain injuries based on the same data. The test is called  $\chi^2$-test.

**Test Steps**

- $H_0$: Suffering a head injury is not associated with wearing a helmet
- $H_a$: There is an association between wearing a helmet and suffering a head injury
- Specify the alpha level of the test
- Rejection rule (two-sided only):
Reject Ho if the $\chi^2$ statistic is too large (see discussion below)



Implication of the null hypothesis
---
- the head-injury rate: $\hat p = 235/793 = 0.2963$;
- observations
    - The expected number of injured protected cyclists is $0.2963\times147 = 43.6$
    -  the number of injured unprotected cyclists should be $0.2963\times646 = 191.4$
-  Let $E_i$ is the expected number of head injuries, $O_i$ is the observed number, $r$ is the number of rows in the table, and $c$ is the number of columns. $\chi^2$ statistic is defined as follows:
$$ \chi^2_{(r-1)(c-1)}=\sum_{i}^{rc}\frac{(O_i-E_i)^2}{E_i}$$
here, degree of freedon is 1 in the above example, $(2-1)\times(2-1)=1$.

Statistics
---
Obviously, the shape of pdf of $\chi^2_{df=1}$ with 1 degree of freedom is different from that of others; the  statistics is generally modified (**Yate's correction**) as follows:
\begin{eqnarray}
  \chi^2_1&=&\sum_{i}^{rc}\frac{(|O_i-E_i|-0.5)^2}{E_i}\\
\end{eqnarray}
By this modified formula, this implies  $\chi^2_1=27.27$.

The chi-square test in the previous example is implemented as follows:
1. $H_0$: Suffering a head injury is not associated with wearing a helmet<br>
2. $H_a$: There is an association between wearing a helmet and suffering a head injury<br>
3. $\alpha = 0.05$<br>
4. Rejection rule (two-sided only):
   Reject $H_0$ since the chi-square statistic, 27.27, is higher than $\chi^2_{1,0.05} = 3.84$.

The odds ratio (OR)
---
The chi square test of association answers **only** the question of association. For a further investigation of this hypothesis,  the nature or direction of the association, one must rely on a different test.

<div class="alert alert-warning">
$$\hbox{OR}= \frac
      {{\cal P}(\hbox{Disease | exposed})/[1-{\cal P}(\hbox{Disease | exposed}]}
      {{\cal P}(\hbox{Disease | unexposed})/[1-{\cal P}(\hbox{Disease | unexposed}]}      
$$

- odds of having the disease if exposed is
$$\frac{{\cal P}(\hbox{Disease | exposed})}{1-{\cal P}(\hbox{Disease | exposed})}$$

- odds of having the disease if unexposed is
$$\frac{{\cal P}(\hbox{Disease | unexposed})}{1-{\cal P}(\hbox{Disease | unexposed})}$$

Statistics
---
$$\widehat{\hbox{OR}}=\frac{\frac{a/(a+c)}{c/(a+c)}}{\frac{b/(b+d)}{d/(b+d)}}
     =\frac{ad}{bc}$$
is not Normal distributed; but its logarithm is:
$$Z=\frac{\ln(ad/bc)}{\sqrt{1/a+1/b+1/c+1/d}}$$

Test Hypothesis
---
1. Null $H_0$:<br>
   a) One-sided tests:
      - $H_0:OR\ge1$
      - $H_0:OR\le1$
      
   b) two-sided test:
      - $H_0:OR=1$
2. alternating  $H_a$:<br>
   a) One-sided tests:
      - $H_a:OR<1$
      - $H_a:OR>1$
      
   b) two-sided test:
      - $H_a:OR\ne1$   
3. specify the $\alpha$ level.

4.test statistics, $Z=\frac{\ln(ad/bc)}{\sqrt{1/a+1/b+1/c+1/d}}$<br>
5. Rejection rule:

  a) One-sided alternatives
    -  Reject the null hypothesis if $Z < z_{1âˆ’\alpha}$ 
    -  Reject the null hypothesis if $Z < -z_{1âˆ’\alpha}$ 
    
  b) Two-sided alternative
    - Reject the null hypothesis if $Z > z_{1âˆ’\alpha/2}$ or if $Z < âˆ’z_{1âˆ’\alpha/2}$.

Confidence intervals
---
A $(1 âˆ’ \alpha)$% two-sided confidence interval of the log-odds ratio is given by
$$\left(\ln(\overline{OR})-z_{1-\alpha/2}\sqrt{1/a+1/b+1/c+1/d},
  \ln(\overline{OR}+z_{1-\alpha/2}\sqrt{1/a+1/b+1/c+1/d}
  \right)$$
  and the configence interval is 
$$\left(\exp\left\{\ln(\overline{OR})-z_{1-\alpha/2}\sqrt{1/a+1/b+1/c+1/d}\right\},
  \exp\left\{\ln(\overline{OR}+z_{1-\alpha/2}\sqrt{1/a+1/b+1/c+1/d}\right\}
  \right)$$

The question that naturally arises is whether we should combine the information in those two tables and use all the available data in order to ascertain the effect of coffee on the risk of Myocardial Infarction (MI). However, if the association between coffee and MI is different in the group of smokers compared to the group of non-smokers, then such an analysis would be *inappropriate*.

Strategy
---
1. Analyze the two tables separately. Based on the individual estimates of the odds ratios,

2. Test the hypothesis that the odds ratios in the two subgroups are sufficiently close to each other (they are homogeneous).

$$\left(Y-z_{1-\alpha/2}s.e.(Y), Y+z_{1-\alpha/2}s.e.(Y)\right)$$

Mantel Haenszel test
---
1. $H_0: OR_1 = 1$
2. $H_a: OR_1\ne 1$ (only two-sided alternatives can be accommodated by the chi-square test)
3. The Mantel-Haenszel test statistic:
$$X^2=\frac{\sum_i a_i-\sum_i m_i}{\sum_i\sigma^2_i}\sim\chi^2_1$$4.Reject $H_0$ if $X^2>\chi^2_{1,1-\alpha}$

$a_1 = 1011, m_1 = 981.3, \sigma_1^2 = 29.81, a_2 = 383, m_2 = 358.4, \sigma^2_2 = 37.69$. Thus,
$$X^2_{MH}=\frac{1011+383-981.3-358.4}{29.81+37.69}=43.68>\chi^2_{1,0.95}=3.84$$
