Diagnostic tests
---
Consider the following events: 
- $D$ = “Disease is present”
- $D^C$ = “Disease is absent”
- $T^+$ = “Positive test result (test detects disease)”
- $T^−$ = “Negative test result (test does not detect disease)”



In diagnostic-testing situations, the following “performance parameters” of the diagnostic
procedure under consideration will be available:
- $ {\cal P}(T^+|D)$ = “Sensitivity (true positive rate) of the test”
- $ {\cal P}(T^+|D^C)$ = “Probability of a false positive test result”
- $ {\cal P}(T^+|D)$ = “Probability of a false negative test result”
- $ {\cal P}(T^-|D^C)$ = ”Specificity (or true-negative rate) of the test”

In addition, in order to derive estimates of the P V P, i.e., the predictive value of a positive test 
$$P V P = {\cal P} (D|T^+) $$
(with respest to  the predictive value of a negative test: 
$$P V N = {\cal P} (D^C|T^−) $$
we will need an estimate of the overall probability of disease in the general population. This is called **the prevalence of the disease** ${\cal P}(D)$.

**Goal**: Find ${\cal P}(D|T^+)$ the predictive value of a positive test result (or PVP), that is, find the probability that a subject has the disease given a positive test.

Given ${\cal P}(D)=0.000093$ but not from the estimation; it is reasonable since the samples are not chosen randomly  from thw whole population.

$$(22+8)/(22+8+51+1739)=30/1820\sim0.022$$
- ${\cal P}(T^+|D)$ = sensitivity =$\frac{22}{30}\sim0.7333$
- ${\cal P}(T^-|D)$ = “false negative rate”=$\frac{8}{30}\sim0.2667$
- ${\cal P}(T^+|D^C)$ = “false positive rate” =$\frac{51}{1790}\sim0.0285$
- ${\cal P}(T^-|D^C)$ = specificity =$\frac{1739}{1790}\sim0.9715$


\begin{eqnarray}
{\cal P}(T^+) &=& {\cal P}􏰏(D\cap T^+)\cup(T^+ \cap D^C)􏰐 \\
             &=& {\cal P}(D\cap T^+)+{\cal P}(T^+ \cap D^C)\\
             &=& {\cal P}(T^+|D)P(D) + {\cal P}(T^+|D^C)P(D^C)\\
             &\Downarrow&\\
  {\cal P} (D|T^+)&=& \frac{{\cal P} (D\cap T^+)}{{\cal P}(T^+)}
                   = \frac{{\cal P} (D\cap T^+)/P(D)\cdot P(D)}{{\cal P}(T^+)}\\ 
             &=& \frac{{\cal P} (T^+|D)P(D) }{{\cal P}(T^+|D)P(D) + {\cal P}(T^+|D^C)P(D^C)}
\end{eqnarray}


- PVP = ${\cal P} (D|T^+)=0.7333*0.000093/(0.7333*0.000093+0.0285*(1-0.000093))\sim0.00239$ 
  
  This shows: 
for every 100,000 positive x-rays, only 239 signal true cases of tuberculosis



````r
0.7333*0.000093/(0.7333*0.000093+0.0285*(1-0.000093))
````

As same procedure,
\begin{eqnarray}
{\cal P}(T^-) &=& {\cal P}􏰏(D\cap T^-)\cup(T^- \cap D^C)􏰐 \\
             &=& {\cal P}(D\cap T^-)+{\cal P}(T^- \cap D^C)\\
             &=& {\cal P}(T^-|D)P(D) + {\cal P}(T^-|D^C)P(D^C)\\
             &\Downarrow&\\
  {\cal P} (D^C|T^-)&=& \frac{{\cal P} (D^C\cap T^-)}{{\cal P}(T^-)}
                   = \frac{{\cal P} (D^C\cap T^-)/P(D^C)\cdot P(D^C)}{{\cal P}(T^-)}\\ 
             &=& \frac{{\cal P} (T^-|D^C)P(D^C) }{{\cal P}(T^-|D)P(D) + {\cal P}(T^-|D^C)P(D^C)}
\end{eqnarray}

Define Functions
---
- probDE(x): conditional probability of disease on exposed (unexposed)
- RR(x,y)= probDE(x)/probDE(y)

Note: 
 - %.3f: scientific format with integer.xxx

````r
# define functions, probDE=Prob(D|E),RR(x,y)
probDE <- function(x){
    x[1]/sum(x)
}
RR <- function(x,y){
    cat(sprintf("The RR is %.3f",probDE(x)/probDE(y)))
}
````

$$\hbox{OR}= \frac

6.4 Diagnostic Tests ( 診斷檢定 )
---

|Cutpoint| $D$| $D^C$|
|:---:|:---:|:---:|
|$T^+$|$a$|$c$|
|$T^-$|$b$|$d$|

- Sensitivity ( 敏感度 ) =  (True positive rate):  有病者診斷結果為陽性的機率
$$ {\cal P}(T^+|D)=\frac{a}{a+b}$$

- Specificity ( 特異度 ) = (True negative rate) : 沒病者診斷結果為陰性的機率
$$ {\cal P}(T^-|D^C)=\frac{d}{c+d}$$

記憶口訣
---
- 敏感度: 有病的人, 被檢驗出來真的有病的機率,
- 特異度: 沒病的人, 被檢驗出來也沒有病的機率.

Note
---
The less the threshhold is, the more both the TruePositives and the FalsePositives are.

````r
SumD=sum(D)
SumDC=sum(DC)

TruePositives <- rep(0,len)
FalsePositives <- rep(0,len)
Specificity <- rep(0,len)

for(i in c(1:len)){
   TruePositives[i] <- 1-sum(D[1:i])/SumD
   FalsePositives[i]<-1-sum(DC[1:i])/SumDC
   Specificity[i]<- 1-FalsePositives[i]
}
````

````r
CutoffPoint <- c(1,1.3,1.6,1.9,2.2,2.5,2.8)
Sensitivity <- TruePositives
#sprintf("%.3f",Sensitivity)
tableD <-cbind(CutoffPoint,Sensitivity,Specificity,FalsePositives)
tableD
````

- Positive predictive value (陽性預測值 , predictive value of a positive test)
\begin{eqnarray}
{\cal P}(D|T^+)&=& \frac{{\cal P}(D\cap T^+ )}{{\cal P}(T^+)}\\
               &=&\frac{{\cal P}(T^+ |D){\cal P}(D)}
                  {{\cal P}(T^+ |D){\cal P}(D)+{\cal P}(T^+ |D^C){\cal P}(D^C)}
\end{eqnarray}

- Negative predictive value ( 陰性預測值 , predictive value of a negative test) :
\begin{eqnarray}
{\cal P}(D^C|T^-)&=& \frac{{\cal P}(D^C\cap T^- )}{{\cal P}(T^-)}\\
               &=&\frac{{\cal P}(T^- |D^C){\cal P}(D^C)}
                  {{\cal P}(T^- |D){\cal P}(D)+{\cal P}(T^- |D^C){\cal P}(D^C)}
\end{eqnarray}

記憶口訣
---
- 陽性預測值: 被檢驗有病的人, 真的有病的機率,
- 陰性預測值: 被檢驗沒病的人, 真的沒病的機率.

- ${\cal P}(D^C)= 1 - {\cal P}(D)=0.85$
