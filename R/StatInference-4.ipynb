For testing hypothesis: with $\alpha$-level,
$$H_0:\cdots\\H_a:\cdots$$

 **a)** Type I error: 
  $$ {\cal P}(H_a|H_0)$$
  The $\alpha$ level of the test is the maximum allowed probability of type-I error.
  <br>
 **b)** Type II error: 
  $$ \beta={\cal P}(H_0|H_a)$$
  
 **c)** Power:  the probability that the test will correctly reject the null hypothesis
 $$ \hbox{ power }=1-\beta$$

Consider the following example. U.S. males ages 20-24 have mean cholesterol level $\mu$ = 180 mg/ml. By comparison, we can assume that the mean cholesterol level in the overall population of males 20-74 is higher. Thus, we are carrying out the one-sided test of hypothesis:
```
H0 : μ≤180 mg/100 ml 
Ha : μ>180 mg/100ml
```
If we collect a sample of n = 25 U.S. males ages 20-74 and measure their cholesterol level, we can
use $\bar X_n$, the sample mean (mean cholesterol level in our sample) to make inferences about the (20-74
year-old male) population mean. The distribution of $\bar X_n$ is normal, with mean μ = 180 mg/100 ml
(by the assumption implicit in the null hypothesis), and standard deviation $\sigma=46/\sqrt{25}$ mg/ml (by
 the central limit theorem).

If $\alpha$=0.05 was chosen,
- cut-off point, $\bar x_n $: we will reject the null hypothesis, and decide that the 20-74 population mean cholesterol level is higher than the 20-24 year old population if
\begin{eqnarray*}
  0.05 & = &   {\cal P}(H_a|H_0)\\
         & = & {\cal P}(X_n \ge \bar{x}_n|\mu=180)\\
         & = & {\cal P} \left( Z=\frac{\bar{X}_n -{\mu}}{\sigma /\sqrt{n}}  \ge \frac{\bar{x}_n -{\mu}}{\sigma /\sqrt{n}}  \right)\\
  \Rightarrow &  & \frac{\bar{x}_n -{180}}{46 / \sqrt{25}} = z_{0.95}=1.64 \\
  \Rightarrow &&\bar X_n \ge \bar{x}_n =195.1 
\end{eqnarray*}
- What if the ”true” mean cholesterol of 20-74 year-olds is μ1 = 211 mg/dL? This causes the type II error, $\beta$,
\begin{eqnarray*}
  {\cal P}(H_0|H_a)
         &=& {\cal P} \left( \bar{X}_n\le 195.1|\mu=\mu_1=211\right)\\
         & = & {\cal P} \left( \frac{\bar{X}_n -211}{46 /\sqrt{25}}  \le \frac{195.1 -211}{46 /\sqrt{25}}=-1.73  \right)\\
  \Rightarrow &  & \beta=0.042\\
\end{eqnarray*}
- power: $1-\beta=0.958$

Sample Size Estimation
---
The more sample size increases, the further the null and alternative distributions seperate with each other , as above pictures shown. 

To determine the required **sample size**, we need the following information:

1. Null and alternative means ($\mu_1,\mu_2$)

2. Standard deviation (or estimate of variability), sd

3. Alpha level of the test ($\alpha$)

4. Desired power ($1-\beta$)

Example Continued
---
Given $\alpha,\beta$, how many of observations should be taken?
For the case:

$$H_0: \mu\le180\\
  H_a: \mu>180$$
and $\mu_0=180$ mg/dL and $\mu_1=211$ mg/dL.  Then
\begin{eqnarray}
     \alpha=0.01 &=& {\cal P}(H_a|H_0)\\
                &=& {\cal P} \left( Z\le\frac{\bar{x}_n -180}{46 /\sqrt{n}}  = z_{0.99}({\color{red} {=z_{1-\alpha}}})=2.32 \right)\\
      \Longrightarrow &&        \bar{x}_n=180+2.32 \frac{46} {\sqrt{n}}\\
      \beta=0.05 &=& {\cal P}(H_0|H_a)\\
                &=& {\cal P} \left( Z\le\frac{\bar{x}_n -211}{46 /\sqrt{n}}  \le z_{0.05}({\color{green}{=z_{\beta}} })=-1.645 \right)\\
      \Longrightarrow &&        \bar{x}_n=211-1.645 \frac{46} {\sqrt{n}}\\
       \Longrightarrow &&        180+2.32 \frac{46} {\sqrt{n}}=211-1.645 \frac{46} {\sqrt{n}}\\
        \Longrightarrow && n\approx34.6\approx35
\end{eqnarray}

Conclusion
---
To be assured of a 95% chance of detecting differences in cholesterol level between 20-74 and 20-24 year-old males (power) when carrying out the test at a 1% $\alpha$ level, we would need about 35 20-74 year-old males.

Note
---
For general $\alpha,\beta$,
\begin{eqnarray}
 \bar x &=& \mu_0+ {\color{red}{z_{1-\alpha}}} \frac{\sigma} {\sqrt{n}}\\
        &=& \mu_1+ {\color{green}{z_{\beta}}} \frac{\sigma} {\sqrt{n}}\\
 n&=&\left( \frac{ {\color{red}{z_{1-\alpha}}}-{\color{green}{z_\beta}}}{\mu_1-\mu_0}\sigma \right)^2 \\
\end{eqnarray} 

The two (independent) sample case
---
Consider the null hypothesis is $H_0 : \mu_1 = \mu_2$ (or $\delta=\mu_1 - \mu_2=0$).
Previously, we have known that Power and sample-size calculations are based on the distribution of the sample difference of the two means $\bar d = \bar X_1 − \bar X_2$ under some a priori assumptions.

The distribution of the sample difference of two means, assuming two equal-size $n_1 = n_2 = n$, 
(say) independent samples and known and equal variances ($\sigma_1 = \sigma_2 = s$) is $\bar d \sim N(\delta,\sigma_\delta)$, where
$$\sigma_\delta^2=\sigma^2\left( \frac{1}{n_1}+\frac{1}{n_2}\right)=2\sigma^2/n$$

Cholesterol Example
---
Assumption: $n_1=n_2=n$.

If the mean cholesterol among 20-24 year-olds were also *unknown*, while the standard deviation in both groups were $\sigma$ = 46 mg/ml, two samples would be collected. One from a group of 20-24 year-olds and one from a group of 20-74 y ear-olds. 

Under the null $H_0 : \delta=0$ 
 
distribution of $\bar d ∼ N (0, σ^2_{\bar d})$ with $\sigma_{\bar d} = σ{\sqrt{2/n}}=46\cdot\sqrt{2/25}$ = 13.01 mg/dL. 
The procedure of esitmating the sample size as previous example would be as follows:

1. $H_0$: d = 0

2. $H_a$: d = 31 mg/ml (corresponding to the situation where μ1 = 180 mg/ml and μ2 = 211
mg/ml)
3. $\alpha=0.01\to z_{1-\alpha}=2.32$;

4. Power = $1 − \beta = 0.95\to z_\beta=-1.645$.

By the formula avove,
$$n'= \left( \frac{z_{1-\alpha}-z_\beta}{\mu_1-\mu_0}\sqrt2\sigma \right)^2=2\left( \frac{z_{1-\alpha}-z_\beta}{\mu_1-\mu_0}\sigma \right)^2\approx70$$
roughly double the size for one-sample case, require 70 cases for per group.

Power calculations when testing a single proportion
---

When determin- ing power or sample size for the comparison of a proportion (against a known population value), we make use of the normal approximation of the binomial distribution. That is, we use the fact that, at least for large n,
statistic:
$$ Z=\frac{\hat p -p }{\sqrt{p(1-p)/n}}$$

example Five-year survival rate of lung-cancer patients
---
Suppose that we are testing the hypothesis that the 5-year survival of lung-cancer patients under 40 is 8.2%, equal to 5-year survival rates of lung-cancer patients over 40 years-old.

$$H_0: p_0\le0.082\\
  H_a: p_0>0.082$$

- $n=52$
- Under $H_0$, $\sigma_{\hat p}=\sqrt{p_0(1-p_0)/n}=\sqrt{0.082(1-0.082)/52}=0.38$
- Under $H_1$, $\sigma_{\hat p}=\sqrt{p_a(1-p_a)/n}=\sqrt{0.2(1-0.2)/52}=0.055$

if the 5-year survival among younger patients is as high as 20%
(i.e., $p_a$ = 0.200)


Reject $H_0$
---
if 
\begin{eqnarray}
    \alpha &=& {\cal P}\left( Z=\frac{\hat p-p_0}{\sqrt{p_0(1-p_0)/n}}\ge z_{1-\alpha} \right) \\
           \Longrightarrow && \hat p\ge p_0+z_{1-\alpha}\sqrt{p_0(1-p_0)/n}
\end{eqnarray}

For $\alpha=0.05$,
$$ \hat p\ge p_0+z_{1-\alpha}\sqrt{p_0(1-p_0)/n}=0.082+1.645\sqrt{0.082(1-0.082)/52}\approx0.145$$

````r
options(repr.plot.width=7)
options(repr.plot.height=4)
options(warn=-1)
oldpar <- par(mar = c(5, 6, 4, 2) + 0.1) 

x <- seq(-0.0, 0.4, 0.005) 
n<-52
p1<-0.082
p2<-0.2
sd1<-sqrt(p1*(1-p1)/n)
sd2<-sqrt(p2*(1-p2)/n)
fx1 <-dnorm(x,mean=p1,sd=sd1)

plot(x, fx1, type = "l", col='blue',ylim=c(-0.1,11),xlim=c(-0.1,0.4),lwd = 2,xlab="",ylab="",
     main=expression(H[0]:p[0]<=0.082)) 
par(new=TRUE)
x <- seq(-0.0, 0.4, 0.005) 

fx2 <-dnorm(x,mean=p2,sd=sd2)
plot(x, fx2, type = "l", ylim=c(-0.1,11),xlim=c(-0.1,0.4),col='brown',lwd = 1,xlab="",ylab="",axes=F) 

abline(v=0.082, col="gray");
abline(v=0.2, col="gray");
abline(v=0.145, col="red",lwd=2);
#text(0.145,1,expression(bar(p)==0.145))
text(0.145,1,expression(p[0]+z[1-alpha/2]*sqrt(p[0](1-p[0])/sqrt(n))==0.145),col='red')

text(0.08,6,expression(p==p[0]))
text(0.2,6,expression(p==p[a]))
text(0.31,7,expression(alpha==0.05))
text(0.3,8,expression(n==52))
text(0.3,10,"Rejection Region")
text(0.,10,"Acceptance Region")
````

Under $H_1$, $n=52,\alpha=0.05$ known:
- $p_a=0.2\to \hat p\sim N(p_a,p_a(1-p_a)/n)=N(0.2,0.055^2)$

\begin{eqnarray}
    \beta &=& {\cal P}\left( H_0|H_a \right) \\
          &=& {\cal P}\left(\hat p< p_0+z_{1-\alpha}\sqrt{p_0(1-p_0)/n} \right) \\
          &=& {\cal P}\left( {\hat p-p_a< (p_0-p_a)}+ z_{1-\alpha}{\sqrt{p_0(1-p_0)/n}} \right) \\
          &=& {\cal P}\left( Z=\frac{\hat p-p_a}{\sqrt{p_a(1-p_a)/n}}\le
          \frac{(p_0-p_a)+ z_{1-\alpha}{\sqrt{p_0(1-p_0)/n}} }{\sqrt{p_a(1-p_a)/n}}\right) \\
          &=& {\cal P}( Z\le z_\beta) \\
           \Longrightarrow && z_\beta =\frac{(p_0-p_a)+ z_{1-\alpha}{\sqrt{p_0(1-p_0)/n}} }{\sqrt{p_a(1-p_a)/n}}
\end{eqnarray}

Here, $\alpha$=0.05, $z_\beta\approx-1.000$ and type II error, $\beta$, is approximate 0.242.


Estimaton of size samples, $n$
---

If the power, $1-\beta$, is 95% and the $\alpha$ level of the test is 1%, then $z_\beta=-1.645,z_{1-\alpha}=2.32$; the required sample size is

\begin{eqnarray}
-1.645=z_\beta &=&\frac{(p_0-p_a)+ z_{1-\alpha}{\sqrt{p_0(1-p_0)/n}} }{\sqrt{p_a(1-p_a)/n}}\\
\to p_a-p_0&=&\left[{-z_\beta{\sqrt{p_a(1-p_a)/n}+z_{1-\alpha}{\sqrt{p_0(1-p_0)/n}}}}\right]^2\\
n&=&\left[ \frac{1.645{\sqrt{p_a(1-p_a)}+2.32{\sqrt{p_0(1-p_0)}}}}{p_a-p_0}\right]^2\\
n&\approx&121
\end{eqnarray}

Conclusion
---
- Estimation of size, given $\alpha,\beta$ under $H_0:p\le p_0$ and $p_a$:
$$n=\left[ \frac{-z_\beta{\sqrt{p_a(1-p_a)}+z_{1-\alpha}{\sqrt{p_0(1-p_0)}}}}{p_a-p_0}\right]^2$$
- Estimation of size, given $\alpha,\beta$ under $H_0:p= p_0$ and $p_a$:
$$n=\left[ \frac{-z_\beta{\sqrt{p_a(1-p_a)}+z_{1-\alpha/2}{\sqrt{p_0(1-p_0)}}}}{p_a-p_0}\right]^2$$
