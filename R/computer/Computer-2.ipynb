繪圖1 (圖餅圖) 教育程度及抽菸。


````r
# use Mac font and surpress the warnings
#par(family="STKaiti") 
#par(family="STHeiti") 
#par(family="STSong") 
par(family="LiHei Pro") 
#par(family="LiSong Pro") 


options(warn=-1)

table1<-table(data1$smoke)
# sum up the number of smoking persons 
smokey <- (sum(data1$smoke))
lbls<-c("不抽菸","抽菸")

# create the length array of (no smoke, smoke)
pct <- c(length(data1$smoke)-smokey,smokey)

# change length array into portion array
pct <- paste(pct*100/length(data1$smoke),"%",sep="") 
lbls <- paste(lbls, pct) 

pie(table1,labels = lbls, main="抽菸比例")
````

````r
par(family="STKaiti") 
options(warn=-1)
# total number of attendants
EDUnum <- length(data1$EDU)
#par(xpd=TRUE)


# calculate number in each category
EDUarr <-c(sum(data1$EDU==0),sum(data1$EDU==1),sum(data1$EDU==2),sum(data1$EDU==3),sum(data1$EDU==4),
           sum(data1$EDU==5),sum(data1$EDU==6),sum(data1$EDU==7))
#lbls<-c("Edu Level 0\n","Edu Level 1\n","Edu Level 2\n","Edu Level 3\n","Edu Level 4\n",
#        "Edu Level 5: ","Edu Level 6: ","Edu Level 7: ")
lbls<-c("無\n","國小肄業\n","國小畢業\n","國中畢業\n","高中職\n",
        "專科: ","大學: ","研究所或以上 ")


# caculate the portion of each category
pct <- round(EDUarr*10000/EDUnum)/100
pct <- paste(pct,"%",sep="") 
lbls <- paste(lbls, pct) 


table2<-table(data1$EDU)
pie(table2, labels = lbls,main="教育程度",radius=0.8,cex=0.7)
````
