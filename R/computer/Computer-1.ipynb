```r
# case one: cypress pine data
x <- c(20, 36, 38, 39, 48, 56, 68, 73, 73, 75, 80, 122)
```

```r
x.b <- boxplot(x)
```

```r
x.b
```

```r
x.b$stats
```

```r
Q1 <- x.b$stats[2]
Q3 <- x.b$stats[4]
Q1;Q3
```

```r
IQR<-Q3-Q1
IQR
```

```r
LAV <- Q1 - 1.5 * IQR
UAV <- Q3 + 1.5 * IQR
LAV;UAV
```

```r
boxplot(x, boxwex = 0.4, staplewex = 0.2, ylim = c(-20, 150), ylab="Y-axis")
lines(c(0,6), c(UAV, UAV), lty = 3)
lines(c(0,6), c(LAV, LAV), lty = 3)
text(3, 0.5, label = "UAV")
```
