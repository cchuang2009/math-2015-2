$t$ test for equality of $k$ group means
---
1. $H_0 :\mu_1=\mu_2=\mu_3$ versus $H_a:$ at least one pair is not equal;
2. Perform $g=3\times2=6$ two-sample t tests:
   - $H_{0,l} :\mu_i =\mu_j, \quad i,j$ are two of the  $k=3$ groups,and and $l=1,...,g$.
   - $H_{a,l}:\mu_i \ne\mu_j$;
   - significance is $(1-\alpha^*)$%-level;
   - The test statistic is
     $$ T=\frac{\bar x_i-\bar x_j}{\sqrt{s_{p,ij}^2 \left( \frac{1}{n_i}+\frac{1}{n_j}\right)}}$$
     where $s_{p,ij}^2=\frac{(n_1-1)s_i^2+(n_2-1)s_j^2}{n_i+n_j-2}$;
   - Pair-wise test rejection rule: Reject $H_{0,l}$ if $T > t_{n_i+n_j−2;1-\alpha^∗/2}$,
     or if $T < t_{n_i+n_j−2;\alpha^∗/2}$.  

3.  Rejection rule (of the overall test): Reject $H_0$ if any of the $g$ pair-wise tests rejects its null hypothesis $H_{0,l}$. Otherwise, do not reject $H_0$.    

What $\alpha^*$ is 
---
Consider $k=3$, (then $l=6$), and let 
  - event $A$ = “The overall test correctly rejects $H_0$”,
  - and $A_l$ = “Pair-wise test l correctly rejects $H_{0,l}$”,
Then 
$$1-\alpha^*={\cal P}(A)={\cal P}(A_1\cap\cdots\cap A_6)={\cal P}(A_1)\cdots{\cal P}(A_6)=(1-\alpha)^6\le1-\alpha$$
if $\{A_i\}_i$ are independent, i.e the significance level of the overall test can be much smaller than anticipated. 

Sources of Variation
---

The total variability in X considered here is:

\begin{eqnarray}
    S_X&=&\sum_{i=1}^{k}\sum_{j=1}^{n_i}\left(X_{ij}-{\bar X}\right)^2\\
       &=&\sum_{i=1}^{k}\sum_{j=1}^{n_i}\left[(X_{ij}-\bar X_i)+
                         (\bar X_i-{\bar X})\right]^2\\
       &=&\sum_{i=1}^{k}\sum_{j=1}^{n_i}(X_{ij}-\bar X_i)^2
          +\not{\color{red}2\cdot\sum_{i=1}^{k}\sum_{j=1}^{n_i}\cdots}
          +\sum_{i=1}^{k}\sum_{j=1}^{n_i}(\bar X_i-{\bar X})^2\\
       &=&\sum_{i=1}^{k}\sum_{j=1}^{n_i}(X_{ij}-\bar X_i)^2
          +\sum_{i=1}^{k}\sum_{j=1}^{n_i}(\bar X_i-{\bar X})^2\\
       &=& \sum_{i=1}^{k}(n_i-1)s_i^2+\sum_{i=1}^{k}{n_i}(\bar X_i-{\bar X})^2\\   
       &=& {SS_w}+ {SS_b}     
\end{eqnarray}
where 
  - $N=\sum_in_i$, $\bar X_i =\frac{\sum\limits_{j=1}^{n_i} X_{ij}}{n_i}$, 
       ${\bar X}=\frac{1}{N}\sum_{i=1}^k\sum_{j=1}^{n_i}X_{ij}=\frac{\sum_{i=1}^kn_i\bar X_i}{N}$
  - $SS_{\color{red}w}$: Variability due to differences <font color="red">*w*</font>ithin each of the k groups;
  - $SS_{\color{red}b}$: Variability due to differences <font color="red">*b*</font>etween each of the k groups.
  

Note
---
the sample correlation between diferences of within and between groups is zero since:

\begin{eqnarray}
   \sum_{i=1}^{k}\sum_{j=1}^{n_i}(X_{ij}-\bar X_i)(\bar X_i-\bar{\bar X})
      &=& \sum_{i=1}^{k}(\bar X_i{\bar X})\sum_{j=1}^{n_i}(X_{ij}-\bar X_i)\\
      &=& \sum_{i=1}^{k}(\bar X_i-{\bar X})(n_i\bar X_i-n_i\cdot\bar X_i)=0
\end{eqnarray}

Test Statistics
---
- Variability within each of the k groups, which to a certain extent is inherent. This is estimated by:
  $$ s_W^2=\frac{SS_W}{N-k}=\frac{\sum_{i=1}^{k}(n_i-1)s_i^2}{N-k}$$
-   Variability between the groups is estimated by
  $$ s_B^2=\frac{SS_W}{k-1}=\frac{\sum_{i=1}^{k}{n_i}(\bar X_i-{\bar X})^2}{k-1}$$
-  $F_{k-1,N-k}=s_B^2/s_W^2$

Test Statistics

- Variability within each of the k groups:
  $$ s_W^2=\frac{SS_W}{N-k}=\frac{(21-1)(0.496)^2+(16-1)(0.523)^2+(23-1)(0.498)^2}{21+16+23-3}=0.254$$

Analysis of Variance
---
Assumptions
- The $k$ groups are mutually independent;
- The $k$ group variances are equal, i.e., $\sigma^2_i=\sigma^2$ for $i=1,\cdots,k$;
- The populations are approximately normal

Post-hoc Tests
---
Even though the analysis of variance technique addresses the question of equality of the
group means, there are many instances where we would like to know which subgroups are different. Recall that it suffices for only two groups to be different for the F test to reject the null hypothesis. Normally these tests will be carried out after the null hypothesis has been rejected. This is why they are called post-hoc tests.

Most of these tests perform all pair-wise comparisons between the k group means, ad- justing the p value for the fact that multiple comparisons are carried out.


Note 
---
It is not meaningful to carry out such tests if the overall hypothesis has not been re- jected. In some borderline cases, one or more pair-wise comparison may appear statistically significant, even though the overall hypothesis has not been rejected by the F test.

The Bonferroni test
---
The Bonferroni post-hoc test proceeds as follows:
1. Carry out all pair-wise tests $l$ such that $H_0,l : \mu_i = \mu_j , i, j$ are two of the $k$ groups, and $l = 1,...,g$.
2. $H_{a,l} :\mu_i\ne\mu_j$;
3. The significance level is $(1 − α^∗)$%, where $α^∗ = \alpha/g$ approximately.
4. The test statistic is $T = \frac{\bar x_i−\bar x_j}{s_W^2\sqrt{\frac{1}{n_i}+\frac{1}{n_i}}}$ where $s_W^2$ is the variability within groups taken
from the ANOVA table.
(**Note** We take advantage of information derived from all k groups in calculating the pooled estimate of the variance.)
5. Rejection rule: Reject $H_{0,l}$ if $T > t_{n−k;α^∗/2}$ or if $T < −t_{n−k;α^∗/2}$ (notice the degrees of freedom).
